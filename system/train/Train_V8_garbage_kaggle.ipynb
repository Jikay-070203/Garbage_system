## Training YOLO v8m model with Custom data

```python
import os
import zipfile

# Link t·∫£i dataset t·ª´ Roboflow
dataset_url = "https://app.roboflow.com/ds/GLwDSwKIyH?key=XSaB5aNeX0"

# ƒê∆∞·ªùng d·∫´n l∆∞u dataset trong Kaggle
save_dir = '/kaggle/working/dataset'

# T·∫°o th∆∞ m·ª•c n·∫øu n√≥ ch∆∞a t·ªìn t·∫°i
os.makedirs(save_dir, exist_ok=True)

# T·∫£i dataset
try:
    # S·ª≠ d·ª•ng wget ƒë·ªÉ t·∫£i xu·ªëng file
    os.system(f"wget -O {save_dir}/dataset.zip '{dataset_url}'")

    # Gi·∫£i n√©n file
    with zipfile.ZipFile(f"{save_dir}/dataset.zip", "r") as zip_ref:
        zip_ref.extractall(save_dir)

    print("Dataset ƒë√£ ƒë∆∞·ª£c t·∫£i xu·ªëng v√† gi·∫£i n√©n th√†nh c√¥ng!")

except Exception as e:
    print(f"L·ªói khi t·∫£i xu·ªëng ho·∫∑c gi·∫£i n√©n dataset: {e}")
```

```python
#Di chuy·ªÉn file d·ªçn d·∫πp data
import os
import shutil

# ƒê∆∞·ªùng d·∫´n th∆∞ m·ª•c ch·ª©a file ZIP
source_dir = '/kaggle/working/dataset'  # Thay th·∫ø b·∫±ng ƒë∆∞·ªùng d·∫´n th·ª±c t·∫ø

# T√™n file ZIP
zip_file_name = 'dataset.zip'

# ƒê∆∞·ªùng d·∫´n ƒë·∫ßy ƒë·ªß ƒë·∫øn file ZIP trong th∆∞ m·ª•c ngu·ªìn
source_file_path = os.path.join(source_dir, zip_file_name)

# ƒê∆∞·ªùng d·∫´n ƒë√≠ch (th∆∞ m·ª•c /kaggle/working/dataset)
destination_dir = '/kaggle/working/'

# T·∫°o th∆∞ m·ª•c ƒë√≠ch n·∫øu n√≥ ch∆∞a t·ªìn t·∫°i
os.makedirs(destination_dir, exist_ok=True)

# ƒê∆∞·ªùng d·∫´n ƒë·∫ßy ƒë·ªß ƒë·∫øn file ZIP trong th∆∞ m·ª•c ƒë√≠ch
destination_file_path = os.path.join(destination_dir, zip_file_name)

# Di chuy·ªÉn file ZIP
try:
    shutil.move(source_file_path, destination_file_path)
    print(f"File '{zip_file_name}' ƒë√£ ƒë∆∞·ª£c di chuy·ªÉn t·ª´ '{source_dir}' ƒë·∫øn '{destination_dir}' th√†nh c√¥ng!")
except FileNotFoundError:
    print(f"L·ªói: Kh√¥ng t√¨m th·∫•y file '{zip_file_name}' trong th∆∞ m·ª•c '{source_dir}'.")
except Exception as e:
    print(f"L·ªói khi di chuy·ªÉn file: {e}")
```
